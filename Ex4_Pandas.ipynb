```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
```

```python
dados = pd.read_csv(url, sep=';')
dados.head()
```

# Criando colunas numéricas

```python
dados['Custo_Mensal'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Custo_Anual'] = dados['Custo_Mensal']*12 + dados['IPTU']

dados.head()
```

# Criando uma coluna categórica

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' Quartos(s) ' + ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem '

dados.head()
```

# Criando uma coluna binária

```python
dados['Possui_Suites'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False)
pd.read_csv('dados_completos_dev.csv')
```
