# 2. Análise Exploratória dos Dados

## 1) Quais os valores médios de aluguel por tipo de imóvel?

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='purple')
```

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = [
'Conjunto Comercial/Sala',
'Prédio Inteiro', 'Loja/Salão',
'Galpão/Depósito/Armazém',
'Casa Comercial', 'Terreno Padrão',
'Loja Shopping/ Ct Comercial',
'Box/Garagem', 'Chácara',
'Loteamento/Condomínio', 'Sítio',
'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
df_imoveis_residenciais = dados.query('@imoveis_comerciais not in Tipo')
df_imoveis_residenciais.head()
```

```python
df_imoveis_residenciais.Tipo.unique()
```

```python
df_preco_tipo = df_imoveis_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14, 10), color='purple')
```

## 2) Qual o percentual de cada tipo de imóvel na nossa base de dados?

```python
df_imoveis_residenciais.Tipo.value_counts(normalize=True)
```

```python
df_imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
df_percentual_tipos = df_imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_percentual_tipos.plot(kind='bar', figsize=(14,10), color='purple', xlabel='Tipos', ylabel='Percentual')
```

Os imóveis do tipo apartamento representam mais de 80% dos imóveis cadastrados, dessa forma, é razoável manter apenas imóveis desse tipo no dataframe.

```python
df = df_imoveis_residenciais.query('Tipo == "Apartamento"')
df.head()
```

1) Calcular a média de quartos por apartamento;

```python
df['Quartos'].mean()
```

2) Conferir quantos bairros únicos existem na nossa base de dados;

```python
df['Bairro'].unique()
```

3) Analisar quais bairros possuem a média de valor de aluguel mais elevadas;

```python
df.groupby('Bairro').mean('Valor').sort_values('Valor', ascending=False).head()
```

```python
bairros_caros = df.groupby('Bairro')['Valor'].mean('Valor').sort_values(ascending=False).to_frame().head()
bairros_caros
```

4) Criar um gráfico de barras horizontais que apresente os 5 bairros com as médias de valores de aluguel mais elevadas.

```python
bairros_caros.plot(kind='barh', figsize=(14,10), color='red', xlabel='Valor Médio do Aluguel', ylabel='Bairros')
```

# Aplicando filtros

## 1. Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200;

```python
selecao1 = (df['Quartos'] == 1) & (df['Valor'] < 1200)
df[selecao1]
```

```python
df_filtro1 = df[selecao1]
```

## 2. Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m².

```python
selecao2 = (df['Quartos'] == 2) & (df['Valor'] < 3000) & (df['Area'] > 70)
df[selecao2]
```

```python
df_filtro2 = df[selecao2]
```
